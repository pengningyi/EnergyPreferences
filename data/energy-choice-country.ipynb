```python
df = df.groupby(by=["ISO2","Energy_Type"]).sum()
df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
df = df.dropna(how="any")
df.head()
```

```python
import pandas as pd

def reshape_energy_data(df):
    # 1. 识别年份列
    year_columns = [col for col in df.columns if col.startswith('F') and col[1:].isdigit()]
    
    # 2. 使用melt函数将宽格式转换为长格式
    melted_df = pd.melt(
        df,
        # 保持country和energy_type作为标识符
        id_vars=['ISO2', 'Energy_Type'],
        # 年份列作为变量列
        value_vars=year_columns,
        # 设置新的列名
        var_name='year',
        value_name='value'
    )
    
    # 3. 清理年份列，移除'F'前缀
    melted_df['year'] = melted_df['year'].str[1:].astype(int)
    
    # 4. 将energy_type从indexISO2Energy_Type中提取出来
    melted_df['energy_type'] = melted_df['Energy_Type'].str.split('_').str[-1]
    
    # 5. 删除原始的indexISO2Energy_Type列
    melted_df = melted_df.drop('Energy_Type', axis=1)
    
    # 6. 重新排序列
    result = melted_df[['year', 'value', 'ISO2', 'energy_type']]
    
    # 7. 按年份和国家排序
    result = result.sort_values(['year', 'ISO2'])
    
    return result

# 使用示例:
# result_df = reshape_energy_data(your_dataframe)
df = reshape_energy_data(df)
df.head()
```

```python
specific_country = "FI"

country_sums = df.groupby("ISO2")["value"].sum()
threshold = country_sums.quantile(0.95)
top_countries = list(country_sums[country_sums > threshold].index)
if specific_country not in top_countries:
    top_countries.append(specific_country)
```

```python
filtered_df = df[df["ISO2"].isin(top_countries)].copy()
filtered_df.head(20)
```

```python
filtered_df.to_csv("timeline_country_energy_choice.csv")
```
