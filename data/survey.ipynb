```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from collections import defaultdict

# 定义映射字典
mapping = {
    'CBCMelanieStudy1_Random1': {'green_energy': {1:3, 2:2, 3:1}, 'cost': { 1:3, 2:3, 3:2,  }},
    'CBCMelanieStudy1_Random2': {'green_energy': { 1:2, 2:2, 3:3,}, 'cost': { 1:1,2:2,3:2, }},
    'CBCMelanieStudy1_Random3': {'green_energy': { 1:2, 2:1, 3:3,}, 'cost': { 1:1, 2:3, 3:1,}},
    
    'CBCMelanieStudy1_Random4': {'green_energy': { 1:1, 2:1, 3:3, }, 'cost': {1:2,2:1,3:2,}},
    'CBCMelanieStudy1_Random5': {'green_energy': { 1:1, 2:3, 3:2,}, 'cost': { 1:2, 2:1, 3:3, }},
    'CBCMelanieStudy1_Random6': {'green_energy': { 1:3, 2:3, 3:1,}, 'cost': { 1:1, 2:3, 3:3, }},
    'CBCMelanieStudy1_Random7': {'green_energy': { 1:2, 2:1, 3:2,}, 'cost': { 1:2, 2:1, 3:1,}},
    
    'CBCMelanieStudy1_Random8': {'green_energy': { 1:3, 2:3, 3:1,}, 'cost': { 1:2, 2:3, 3:3, }},
    'CBCMelanieStudy1_Random9': {'green_energy': { 1:2, 2:1, 3:3, }, 'cost': { 1:2, 2:1, 3:3, }},
    'CBCMelanieStudy1_Random10': {'green_energy': { 1:2, 2:2, 3:1,}, 'cost': { 1:1, 2:3, 3:2,}},
}
```

```python
freq_data = []

# 逐行计算频率
for _, row in df.iterrows():
    row_income = row['Income']
    row_education = row['Education']
    
    for CBC_col, mappings in mapping.items():
        none_col = CBC_col + '_none'
        
        if row[none_col] == 1:
            green_energy_level = mappings['green_energy'][row[CBC_col]]
            cost_level = mappings['cost'][row[CBC_col]]
            
            freq_data.append({
                'income': row_income,
                'education': row_education,
                'green_energy': green_energy_level,
                'cost': cost_level,
                'combination': (green_energy_level, cost_level)
            })

# 转换为 DataFrame
freq_df = pd.DataFrame(freq_data)

# 按 income 和 education 计算 green_energy 和 cost 的单独频率和联合频率
green_energy_freq = freq_df.groupby(['income', 'education', 'green_energy']).size().reset_index(name='green_energy_freq')
cost_freq = freq_df.groupby(['income', 'education', 'cost']).size().reset_index(name='cost_freq')
combination_freq = freq_df.groupby(['income', 'education', 'combination']).size().reset_index(name='combination_freq')


combination_freq
```

```python
# 计算每个education和income类别的总人数
total_by_income = freq_df.groupby('income').size().reset_index(name='total_income')
total_by_education = freq_df.groupby('education').size().reset_index(name='total_education')

total_by_education.set_index("education")

total_by_education[total_by_education["education"] == 1]["total_education"]
total_by_income.set_index("income")
```

```python

# 合并频率表
merged_freq = green_energy_freq.merge(cost_freq, on=['income', 'education'], how='outer')
merged_freq = merged_freq.merge(combination_freq, on=['income', 'education'], how='outer')
merged_freq
```

```python
climate_columns=[ 'ClimateChangeConcern_r1',
 'ClimateChangeConcern_r2',
 'ClimateChangeConcern_r3',
 'ClimateChangeConcern_r4',
 'ClimateChangeConcern_r5',
 'ClimateChangeConcern_r6',
 'ClimateChangeConcern_r7',
 'ClimateChangeConcern_r8',]

climate_df = df[climate_columns]
climate_df
```

```python
questions = [
    "Concern about climate change",
    "Problems balancing sustainability and family needs",
    "Feeling directly affected by climate change",
    "Knowing someone affected by climate change",
    "Noticing changes in important places",
    "Desire to behave more sustainably",
    "Reducing climate-impacting behaviors",
    "Belief in personal impact on climate change"
]


climate_df.columns = questions
```

```python
data_long = pd.melt(
    climate_df,
    var_name='ConcernAspect',  # 新列名，表示关切维度
    value_name='Score'  # 新列名，表示关切评分
)
data_long
```

```python
def categorize_score(score):
    if score == 1:
        return "Not at all"
    elif score == 2:
        return "Slightly"
    elif score == 3:
        return "Somewhat"
    elif score == 4:
        return "Moderately"
    elif score == 5:
        return "Quite a bit"
    elif score == 6:
        return "Very much"
    else:
        return "Extremely"

data_long['Category'] = data_long['Score'].apply(categorize_score)
data_long.to_csv("climate-concerns-score.csv")
```

```python
percentage_data = data_long.groupby(['ConcernAspect', 'Category']).size().groupby(level=0).apply(lambda x: 100 * x / x.sum()).reset_index(name='Percentage')

neither_data = percentage_data[percentage_data['Category'] == "Moderately"].copy()
neither_data['Percentage'] /= 2
neither_data['Category'] = "Moderately Left"

neither_data_right = neither_data.copy()
neither_data_right['Category'] = "Moderately Right"

percentage_data = percentage_data[percentage_data['Category'] != "Moderately"]
percentage_data = pd.concat([percentage_data, neither_data, neither_data_right], ignore_index=True)

percentage_data
```
