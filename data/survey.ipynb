```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter


# 定义映射字典
mapping = {
    'CBCMelanieStudy1_Random1': {'green_energy': {1:3, 2:2, 3:1}, 'cost': { 1:3, 2:3, 3:2,  }},
    'CBCMelanieStudy1_Random2': {'green_energy': { 1:2, 2:2, 3:3,}, 'cost': { 1:1,2:2,3:2, }},
    'CBCMelanieStudy1_Random3': {'green_energy': { 1:2, 2:1, 3:3,}, 'cost': { 1:1, 2:3, 3:1,}},
    
    'CBCMelanieStudy1_Random4': {'green_energy': { 1:1, 2:1, 3:3, }, 'cost': {1:2,2:1,3:2,}},
    'CBCMelanieStudy1_Random5': {'green_energy': { 1:1, 2:3, 3:2,}, 'cost': { 1:2, 2:1, 3:3, }},
    'CBCMelanieStudy1_Random6': {'green_energy': { 1:3, 2:3, 3:1,}, 'cost': { 1:1, 2:3, 3:3, }},
    'CBCMelanieStudy1_Random7': {'green_energy': { 1:2, 2:1, 3:2,}, 'cost': { 1:2, 2:1, 3:1,}},
    
    'CBCMelanieStudy1_Random8': {'green_energy': { 1:3, 2:3, 3:1,}, 'cost': { 1:2, 2:3, 3:3, }},
    'CBCMelanieStudy1_Random9': {'green_energy': { 1:2, 2:1, 3:3, }, 'cost': { 1:2, 2:1, 3:3, }},
    'CBCMelanieStudy1_Random10': {'green_energy': { 1:2, 2:2, 3:1,}, 'cost': { 1:1, 2:3, 3:2,}},
}

# 统计每个 green_energy 和 cost 组合的频率
green_energy_levels = []
cost_levels = []
level_combinations = []

for col_base in mapping.keys():
    # 选择与当前随机列相关的 none 列
    none_col = f"{col_base}_none"
    subset = df[df[none_col] == 1]
    
    # 对每个符合条件的值进行映射并记录
    for _, row in subset.iterrows():
        green_energy = mapping[col_base]['green_energy'][row[col_base]]
        cost = mapping[col_base]['cost'][row[col_base]]
        
        green_energy_levels.append(green_energy)
        cost_levels.append(cost)
        level_combinations.append((green_energy, cost))

# 计算频率
green_energy_freq = Counter(green_energy_levels)
cost_freq = Counter(cost_levels)
combination_freq = Counter(level_combinations)

# 将频率数据存入 DataFrame
freq_df = pd.DataFrame({
    'Green_Energy': [comb[0] for comb in combination_freq.keys()],
    'Cost': [comb[1] for comb in combination_freq.keys()],
    'Frequency': list(combination_freq.values())
})

# 合并频率数据和原始数据
merged_df = freq_df.merge(df[['Income', 'Education']], left_index=True, right_index=True, how='left')
merged_df.to_csv("green_energy_cost_income_edu.csv")

```

```python
# 创建关系图：Green_Energy + Cost 组合与收入
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Income', hue='Green_Energy', data=merged_df, dodge=True)
plt.title('Green Energy Level and Income based on Frequency')
plt.xlabel('Frequency')
plt.ylabel('Income')
plt.legend(title="Green Energy Level")
plt.show()

# 创建关系图：Green_Energy + Cost 组合与教育
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Education', hue='Cost', data=merged_df, dodge=True)
plt.title('Cost Level and Education Level based on Frequency')
plt.xlabel('Frequency')
plt.ylabel('Education Level')
plt.legend(title="Cost Level")
plt.show()
```
