```python
!pip install matplotlib seaborn pandas numpy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")
```

```python
# Sample data for attributes and part-worth utilities
attributes = ['Color', 'Size', 'Price']
attribute_levels = {
    'Color': ['Red', 'Blue', 'Green'],
    'Size': ['Small', 'Medium', 'Large'],
    'Price': ['$10', '$20', '$30']
}

# Simulated part-worth utilities (random values for demo)
part_worths = {
    'Color': [0.3, 0.1, -0.4],
    'Size': [-0.1, 0.2, 0.1],
    'Price': [0.5, 0.0, -0.5]
}

# Simulated attribute importance
importance = {'Color': 35, 'Size': 25, 'Price': 40}
```

```python
# Part-worth utilities visualization
for attribute, levels in attribute_levels.items():
    plt.figure(figsize=(8, 4))
    sns.barplot(x=levels, y=part_worths[attribute], palette="Blues_d")
    plt.title(f'Part-Worth Utilities for {attribute}')
    plt.xlabel(f'{attribute} Levels')
    plt.ylabel('Utility')
    plt.show()
```

```python
# Attribute importance visualization
plt.figure(figsize=(8, 4))
sns.barplot(x=list(importance.keys()), y=list(importance.values()), palette="Greens_d")
plt.title('Relative Importance of Attributes')
plt.xlabel('Attributes')
plt.ylabel('Importance (%)')
plt.show()
```

```python
# Simulated preference share data
products = ['Product A', 'Product B', 'Product C']
preference_shares = [40, 35, 25]  # Example shares

plt.figure(figsize=(8, 4))
sns.barplot(x=products, y=preference_shares, palette="Purples_d")
plt.title('Simulated Market Share of Products')
plt.xlabel('Products')
plt.ylabel('Preference Share (%)')
plt.show()
```

```python
# Combine data for a heatmap visualization
heatmap_data = pd.DataFrame(part_worths, index=attribute_levels['Color'])

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', center=0)
plt.title('Heatmap of Part-Worth Utilities')
plt.xlabel('Attributes')
plt.ylabel('Attribute Levels')
plt.show()
```

```python
# Simulate individual-level part-worths for one attribute
individual_data = np.random.normal(loc=0, scale=0.2, size=(100, len(attribute_levels['Color'])))

# Convert to DataFrame for visualization
df_individuals = pd.DataFrame(individual_data, columns=attribute_levels['Color'])

plt.figure(figsize=(10, 6))
sns.violinplot(data=df_individuals, palette="Pastel1")
plt.title('Distribution of Individual Part-Worths for "Color" Levels')
plt.xlabel('Color Levels')
plt.ylabel('Part-Worth Utility')
plt.show()
```
