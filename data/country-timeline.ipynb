```python
df = pd.read_csv("./Renewable_Energy.csv")
df.columns
```

```python
import pandas as pd

# 假设你的 DataFrame 名为 df
# 1. 将年份列筛选出来
year_columns = [col for col in df.columns if col.startswith('F')]

# 2. 使用 pd.melt 将这些列转化为一列
df_melted = pd.melt(
    df,
    id_vars=[col for col in df.columns if col not in year_columns],  # 保留非年份的列
    value_vars=year_columns,  # 转换的年份列
    var_name='Year',  # 新列的名称
    value_name='Value'  # 对应的值列名称
)

# 3. 去掉 'F' 前缀（如果需要）
df_melted['Year'] = df_melted['Year'].str.lstrip('F')

# 查看结果
print(df_melted)
```
